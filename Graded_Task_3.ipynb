```python
#Importing Libraries
%matplotlib inline
import numpy as np                   
import matplotlib.pyplot as plt
```

```python
photo_data = plt.imread('drive/MyDrive/IITG.ai/Milky_way.jpg')  #Command to read images
```

```python
#Display image
plt.figure(figsize=(10,10)) 
plt.imshow(photo_data);
```

```python
img_r50 = (photo_data[:, :, 0] < 50)
img = photo_data.copy()
img[img_r50] = 0
plt.imshow(img)
```

```python
from PIL import Image
image = Image.fromarray(img)
image.save('Final_Image.png')  
```
