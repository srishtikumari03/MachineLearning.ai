```python
#Importing Libraries
%matplotlib inline
import numpy as np                   
import matplotlib.pyplot as plt 
```

```python
photo_data = plt.imread('drive/MyDrive/IITG.ai/Milky_way.jpg')  #Command to read images
```

```python
#Display image
plt.figure(figsize=(10,10)) 
plt.imshow(photo_data);     
```

**Required Operation**

```python
#Code Here
img_without_red = photo_data.copy() 
img_without_red[:, :, 0] = 0   #Setting the red channel to 0 to display only green and blue channels.
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(img_without_red)
```
